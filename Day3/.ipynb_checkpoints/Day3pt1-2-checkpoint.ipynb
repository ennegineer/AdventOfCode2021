```python
import os
import csv

file = os.path.join('.', 'input.txt')

binary = []
oxbinary = []
with open(file, 'r') as input:
    csvreader = csv.reader(input)
    
    for row in csvreader:
        binary.append(row)
        oxbinary.append(row)
        print(row)
```

```python
print(oxbinary[0][0][3])
```

```python
gamma = ''
epsilon = ''

# from 0 to end of rows, 0, 0 to 12
for i in range(12):
    zero = 0
    one = 0
    for row in binary:
        if row[0][i] == '0':
            zero = zero + 1
        else:
            one = one + 1
    print(f"{i + 1} zero: {zero} || one: {one}")
    if zero > one:
        print(f"gamma {i + 1} is 0")
        gamma = gamma + '0'
        epsilon = epsilon + '1'
    else:
        print(f"gamma {i + 1} is 1")  
        gamma = gamma + '1'
        epsilon = epsilon + '0'
    print(f" ")
print(f"gamma: {gamma}")
print(f"epsilon: {epsilon}")
```

```python
gammaDec = 1300
epsilDec = 2795

print(gammaDec * epsilDec)
```

```python
oxygen = ''

oxzero = 0
oxone = 0
for row in oxbinary:
    if row[0][0] == '0':
        oxzero = oxzero + 1
    else:
        oxone = oxone + 1
print(f"1 zero: {oxzero} || one: {oxone}")

if oxzero > oxone:
    print(f"oxygen 1 is 0")
    oxygen = oxygen + '0'
else:
    print(f"oxygen 1 is 1")  
    oxygen = oxygen + '1'
print(oxygen)
```
